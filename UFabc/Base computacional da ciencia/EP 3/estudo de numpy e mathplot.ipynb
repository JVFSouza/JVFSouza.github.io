```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10,10,1)
y1 = x*x
y2 = -x*x
y3 = x
plt.plot(x,y1,'b.-',label='trajetória 1')
plt.plot(x,y2,'ro',label='reflexo')
plt.plot(x,y3,'g',label='reta')
plt.title("Gráfico de 3 funções")
plt.grid()
plt.xlabel("tempo")
plt.ylabel("pontos")
% plt.legend()
%% plt.axis('equal')
```
